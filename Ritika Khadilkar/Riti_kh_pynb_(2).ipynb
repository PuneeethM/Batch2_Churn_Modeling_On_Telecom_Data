#Outliers treatment

Boxplot:- Q3+(1.5*IQR)&Q1-(1.5*IQR)

```python
def identify_outliers_iqr(dataset_telecom):
    Q1 = dataset_telecom.quantile(0.25)
    Q3 = dataset_telecom.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (dataset_telecom < lower_bound) | (dataset_telecom > upper_bound)
    return dataset_telecom[outliers.any(axis=1)], lower_bound, upper_bound

outliers, lower_bound, upper_bound = identify_outliers_iqr(dataset_telecom)

# Print the results
print("Identified Outliers using IQR Method:")
print(outliers)
print("\nLower Bound:")
print(lower_bound)
print("\nUpper Bound:")
print(upper_bound)

# Boxplot visualization
plt.figure(figsize=(10, 10))
dataset_telecom.boxplot()
plt.title("Boxplot of Data with Outliers")
dataset_telecom.shape
```

Capping & Flooring

```python
def cap_and_floor(dataset_telecom, lower_percentile=0.05, upper_percentile=0.95):
  # Calculate the lower and upper thresholds
    lower_threshold = dataset_telecom.quantile(lower_percentile)
    upper_threshold = dataset_telecom.quantile(upper_percentile)

    # Apply flooring and capping
    dataset_telecom_capped_floored = dataset_telecom.clip(lower=lower_threshold, upper=upper_threshold, axis=1)

    return dataset_telecom_capped_floored, lower_threshold, upper_threshold
dataset_telecom_capped_floored, lower_threshold, upper_threshold = cap_and_floor(dataset_telecom)

print("Original DataFrame:")
print(dataset_telecom)
print("\nDataFrame after Capping and Flooring:")
print(dataset_telecom_capped_floored)
print("\nLower Thresholds:")
print(lower_threshold)
print("\nUpper Thresholds:")
print(upper_threshold)
# Plot histograms
fig, axes = plt.subplots(2, 1, figsize=(10, 10))

dataset_telecom.hist(bins=30, alpha=0.5, ax=axes[0])
axes[0].set_title("Original Data")

dataset_telecom_capped_floored.hist(bins=30, alpha=0.5, ax=axes[1])
axes[1].set_title("Data after Capping and Flooring")
```

Impute with Mean/Median

```python
def impute_numerical_with_mean_median(dataset_telecom, strategy='mean'):
    numerical_cols = dataset_telecom.select_dtypes(include=['number']).columns
    for col in numerical_cols:
        if strategy == 'mean':
            dataset_telecom[col].fillna(dataset_telecom[col].mean(), inplace=True)
        elif strategy == 'median':
            dataset_telecom[col].fillna(dataset_telecom[col].median(), inplace=True)
    return dataset_telecom
    # Impute numerical columns with mean
dataset_telecom_imputed_mean = impute_numerical_with_mean_median(dataset_telecom, strategy='mean')

print("\nDataFrame after imputing numerical columns with mean:")
print(dataset_telecom_imputed_mean)

# Impute numerical columns with median
dataset_telecom_imputed_median = impute_numerical_with_mean_median(dataset_telecom, strategy='median')

print("\nDataFrame after imputing numerical columns with median:")
print(dataset_telecom_imputed_median)
```

#Removing  the highly coreleated variable

```python
# Removing highly correlated variables

def remove_highly_correlated(dataset_telecom, threshold=0.8):
  # Calculate correlation matrix
    corr_matrix = dataset_telecom.corr().abs()

    # Create a mask to identify highly correlated features
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool_))

    # Find index of feature columns with correlation greater than threshold
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

    # Drop highly correlated features
    dataset_telecom_filtered = dataset_telecom.drop(columns=to_drop)

    return dataset_telecom_filtered, to_drop

# Remove highly correlated variables with threshold of 0.8
dataset_telecom_filtered, dropped_columns = remove_highly_correlated(dataset_telecom, threshold=0.8)

print("\nDataFrame after removing highly correlated variables:")
print(dataset_telecom_filtered)
print("\nColumns removed:")
print(dropped_columns)
```

# Multicollinearity(VIF>5)

```python
def calculate_vif(dataset_telecom):
  vif_data = pd.DataFrame()
  vif_data["Feature"] = dataset_telecom.columns
  vif_data["VIF"] = [variance_inflation_factor(dataset_telecom.values, i) for i in range(dataset_telecom.shape[1])]
  return vif_data
def remove_high_vif(dataset_telecom, threshold=5.0):
     while True:
        vif_data = calculate_vif(dataset_telecom)
        max_vif = vif_data["VIF"].max()
        if max_vif > threshold:
            max_vif_feature = vif_data.loc[vif_data["VIF"].idxmax(), "Feature"]
            print(f"Removing {max_vif_feature} with VIF: {max_vif}")
            dataset_telecom = dataset_telecom.drop(columns=[max_vif_feature])
        else:
            break
            return dataset_telecom
# Remove variables with VIF greater than 5
dataset_telecom_vif_filtered = remove_high_vif(dataset_telecom, threshold=5.0)

print("\nDataFrame after removing variables with high VIF:")
print(dataset_telecom_vif_filtered)


```
